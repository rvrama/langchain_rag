```python
from openai import OpenAI
import time
```

```python
mysk="openapi-api-key"
```

```python
client = OpenAI(api_key=mysk)
```

```python
def wait_for_assistant(thread, run):
    """
        Function to periodically check run status of AI assistant and print run time
    """

    # wait for assistant process prompt
    t0 = time.time()
    while run.status != 'completed':

        # retreive status of run (this might take a few seconds or more)
        run = client.beta.threads.runs.retrieve(
          thread_id=thread.id,
          run_id=run.id
        )

        # wait 0.5 seconds
        time.sleep(0.25)
    dt = time.time() - t0
    print("Elapsed time: " + str(dt) + " seconds")
    
    return run
```

```python
intstructions_string = "RvwGPT, functioning as a virtual writer assistant, using englishmen style book writer's vocabulary to provide a very concise summary and concludes with its signature '–RvwGPT'. \
RvwGPT will tailor the length of its responses to less than 8 words summarizing the viewer's comment, providing concise acknowledgments to brief expressions of gratitude or feedback, \
thus keeping the interaction natural and engaging."

assistant = client.beta.assistants.create(
    name="RvwGPT",
    description="Script writer GPT for User stories or comments",
    instructions=intstructions_string,
    model="gpt-4-0125-preview"
)
print(assistant)
```

```python
# Define an array of user messages
user_messages = [
    "Three men escaped from a high security prison using a tunnel they dug with spoons. In order to avoid detection, they disguised themselves as a rock band and performed at a local concert. Their performance was so convincing that they were able to blend in with the crowd and make their escape unnoticed",
    "A retired teacher decided to open a bookstore in a small town. She filled it with rare first editions and vintage novels. The bookstore became a cultural hub where locals gathered to discuss literature and share stories.",
    "A software engineer developed an AI that could predict plant diseases by analyzing photos. Farmers could now prevent crop loss by catching problems early. The technology was so effective that it spread to agricultural communities worldwide."
]

# Process each user message
for idx, user_message in enumerate(user_messages, 1):
    print(f"\n{'='*60}")
    print(f"Processing Message {idx}")
    print(f"{'='*60}")
    
    # Create a new thread for this message
    thread = client.beta.threads.create()
    
    # Add the user message to the thread
    message = client.beta.threads.messages.create(
        thread_id=thread.id,
        role="user",
        content=user_message
    )
    
    # Send message to assistant to generate a response
    run = client.beta.threads.runs.create(
        thread_id=thread.id,
        assistant_id=assistant.id,
    )
    
    # Wait for assistant to complete
    t0 = time.time()
    while run.status != 'completed':
        run = client.beta.threads.runs.retrieve(
            thread_id=thread.id,
            run_id=run.id
        )
        time.sleep(0.25)
    dt = time.time() - t0
    
    print(f"Elapsed time: {dt:.2f} seconds")
    
    # Retrieve and display the response
    messages = client.beta.threads.messages.list(thread_id=thread.id)
    response_text = messages.data[0].content[0].text.value
    
    print(f"\nInput Message:")
    print(f"{user_message}\n")
    print(f"Assistant Response:")
    print(f"{response_text}")
    print()
```

```python
# wait for assistant process prompt
run = wait_for_assistant(thread, run)

# view run object (in Jupyter Lab)
dict(run)
```
